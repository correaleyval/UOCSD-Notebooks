### Definición de Modelo Lineal de Optimización:

Sea X un espacio vectorial real. Se exige encontrar el óptimo (máximo o mínimo) de una función lineal de <i>f</i> en X, sujeta a cierto número de condiciones lineales del tipo: $g_j(x) \leq b_j ; j = 1, 2, ..., n$

Donde: $b_j \in R$ y las $g_j$ son funciones lineales en $X, j = 1, 2, ..., n$

Aquí las condiciones lineales pueden resolverse usando el Método de los Multiplicadores de Lagrange (estudiado en el Análisis Matemático), pero en la práctica se hace tan complejo el proceso de resolución que es casi imposible su aplicación, esto ha hecho que se estudien como entes matemáticos independientes.


### El problema general de la Programación Lineal

En particular nos interesa el caso en que $X = R^n$ diciendo entonces que estamos en presencia de un problema de Programación Lineal (ppl).

Si $X = R^n$, sabemos que las funciones lienales tienen la forma:

$f(x) = c_1x_1 + c_2x_2 + \cdots + c_nx_n$ donde: $x = (x_1,x_2,\cdots,x_n)$ y $c = (c_1, c_2, \cdots, c_n)$ es un vector fijo de $R^n$ para cada <i>f</i>

En forma matricial  $f(x) = cx^T$, si consideramos los elementos de $R^n$ como vectores fila o matrices de $1 \times n$; pero por comodidad consideremos en lo adelante que x es un vector columna de $n \times 1$, es decir: 

$x = 
\left( \begin{array}{c}
 x_1 \\
 x_2 \\
 \vdots \\
 x_n
\end{array} \right)$

Para facilitar la representación matricial entonces $f(x) = cx$

Teniendo en cuenta esto se puede plantear el problema de la programación lineal de la siguiente forma:
Encontrar el máximo o mínimo de $Z = f(x) = \sum_{j=1}^{n}c_jx_j=cx$ sujeto a las condiciones:

$a_{11} \times x_{1} + a_{12} \times x_2 + \cdots + a_{1n} \times x_n \leq  b_1 \\
a_{21} \times x_{1} + a_{22} \times x_2 + \cdots + a_{2n} \times x_n \leq  b_1 \\
\vdots \\
a_{m1} \times x_{1} + a_{m2} \times x_2 + \cdots + a_{mn} \times x_n \leq  b_1$

donde $c_i, a_{ij}, b_i$; con $i = 1,2, \cdots, m$ y $j = 1,2, \cdots, n$, son números reales dados.

Para abreviar se puede formular el ppl en forma matricial:

Maximizar (o minimizar) $Z = cx$

Sujeto a: $Ax \leq b$ donde $c = \left( c_1, c_2, \cdots, c_n \right)$, $b = \left( b_1, b_2, \cdots, b_m \right)^T$ y 
$A = \left( a_{ij} \right) _{m \times n}$ es la matriz de los $a_{ij}$

```python
### Creacion de matrices en python

import numpy as np

A = np.matrix([[1, 2], 
               [3, 4]])

print(A) # matriz A
print()

print(A.T) # transpuesta de A
print()

print(A @ A.T) # multiplicacion matricial A por A transpuesta
print()

print(A.I) # inversa de la matriz A
print()

print(np.linalg.det(A)) # calculo del determinante
print()

print(A.shape) # ver dimensiones de la matriz
print()
```

```python
# ver graficamente información almacenada en una matriz
import matplotlib.pyplot as plt

plt.plot(A)
plt.show()

plt.plot(A.T)
plt.show()

plt.plot(A.I)
plt.show()
```
